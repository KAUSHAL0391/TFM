Simple Exponential Smoothing

```python
def simple_exponential_smoothing(series, alpha):
    smoothed_series = [series[0]]
    
    for i in range(1, len(series)):
        smoothed_value = alpha * series[i] + (1 - alpha) * smoothed_series[-1]
        smoothed_series.append(smoothed_value)
    
    return smoothed_series
```

```python
alpha = 0.2
smoothed_series = simple_exponential_smoothing(passenger_data, alpha)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(passenger_data, label='Original Series', color='blue')
plt.plot(smoothed_series, label=f'Smoothed Series (Alpha={alpha})', color='red')
plt.title('Original and Smoothed Series (SES)')
plt.xlabel('Year')
plt.ylabel('Passengers')
plt.legend()
plt.grid(True)
plt.show()
```
