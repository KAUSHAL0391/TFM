```python
def simple_moving_average(data, window_size):

    moving_averages = []

    for i in range(len(data) - window_size + 1):
        window = data[i:i+window_size]
        average = sum(window) / window_size
        moving_averages.append(average)

    return moving_averages


input_file ='airline-passengers.csv'
window_size = 12  

data = pd.read_csv(input_file)


sma_values = simple_moving_average(data['Passengers'], window_size)

mae = mean_absolute_error(data['Passengers'][window_size-1:], sma_values)
mse = mean_squared_error(data['Passengers'][window_size-1:], sma_values)
rmse = np.sqrt(mse)

print("MAE:", mae)
print("MSE:", mse)
print("RMSE:", rmse)

```

```python

plt.figure(figsize=(10, 6))
plt.plot(data['Passengers'], label='Original Data')
plt.plot(range(window_size-1, len(data)), sma_values, label='Simple Moving Average', color='green')
plt.xlabel('Day')
plt.ylabel('Price')
plt.title('Simple Moving Average')
plt.legend()
plt.grid(True)
plt.show()
```

```python

def simple_exponential_smoothing(data, alpha):
   
    smoothed_values = [data[0]] 

    for i in range(1, len(data)):
        smoothed_value = alpha * data[i] + (1 - alpha) * smoothed_values[-1]
        smoothed_values.append(smoothed_value)

    return smoothed_values


input_file = "airline-passengers.csv"
alpha = 0.01  

data = pd.read_csv(input_file)


smoothed_values = simple_exponential_smoothing(data['Passengers'], alpha)

mae = mean_absolute_error(data['Passengers'], smoothed_values)
mse = mean_squared_error(data['Passengers'], smoothed_values)
rmse = np.sqrt(mse)

print("MAE:", mae)
print("MSE:", mse)
print("RMSE:", rmse)

plt.figure(figsize=(10, 6))
plt.plot(data['Passengers'], label='Original Data')
plt.plot(smoothed_values, label='Smoothed Data', color='red')
plt.xlabel('Day')
plt.ylabel('Price')
plt.title('Simple Exponential Smoothing')
plt.legend()
plt.grid(True)
plt.show()
```

```python
def holt_winters_smoothing(series, alpha, beta, gamma, seasonal_periods):
   
    level = series[0]
    trend = np.mean([series[i+seasonal_periods] - series[i] for i in range(seasonal_periods)])
    seasonal = [series[i] - level for i in range(seasonal_periods)]

    smoothed_values = [series[0]] 

    for i in range(1, len(series)):
        if i >= seasonal_periods:
            last_level, level = level, alpha * (series[i] - seasonal[i % seasonal_periods]) + (1 - alpha) * (level + trend)
            trend = beta * (level - last_level) + (1 - beta) * trend
            seasonal[i % seasonal_periods] = gamma * (series[i] - level) + (1 - gamma) * seasonal[i % seasonal_periods]
        
        smoothed_values.append(level + trend + seasonal[i % seasonal_periods])

    return smoothed_values


input_file = "airline-passengers.csv"


data = pd.read_csv(input_file)


series = data['Passengers']


alpha = 0.2
beta = 0.2
gamma = 0.2
seasonal_periods = 4


smoothed_values = holt_winters_smoothing(series, alpha, beta, gamma, seasonal_periods)


plt.figure(figsize=(10, 6))
plt.plot(series, label='Original Data')
plt.plot(smoothed_values, label='Smoothed Data', color='red')
plt.xlabel('Day')
plt.ylabel('Price')
plt.title('Holt-Winters Smoothing')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose


input_file = "airline-passengers.csv"


data = pd.read_csv(input_file)


data['date'] = pd.to_datetime(data['date'])


data.set_index('date', inplace=True)


plt.figure(figsize=(12, 6))
plt.plot(data.index, data['Passengers'], label='Original Data')
plt.xlabel('date')
plt.ylabel('Price')
plt.title('Time Series Data')
plt.legend()
plt.grid(True)
plt.show()


decomposition = seasonal_decompose(data['Passengers'], model='additive')


plt.figure(figsize=(12, 8))
plt.subplot(4, 1, 1)
plt.plot(decomposition.trend, label='Trend')
plt.legend()
plt.ylabel('Trend')

plt.subplot(4, 1, 2)
plt.plot(decomposition.seasonal, label='Seasonal')
plt.legend()
plt.ylabel('Seasonal')

plt.subplot(4, 1, 3)
plt.plot(decomposition.resid, label='Residual')
plt.legend()
plt.ylabel('Residual')

plt.subplot(4, 1, 4)
plt.plot(decomposition.observed, label='Observed')
plt.legend()
plt.ylabel('Observed')

plt.xlabel('Date')
plt.suptitle('Decomposition of Time Series Data')
plt.show()
```
